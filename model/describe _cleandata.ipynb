```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r'../clean data/clean_data.csv')
df = df.dropna()

df.info()
```

# CPU

```python
plt.figure(figsize=(20, 6))
sns.countplot(x='CPU', data=df, palette='Set2')
plt.title('Count of CPUs')
plt.xlabel('CPU')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

## RAM

Biểu đồ tương quan heatmap
 

```python
def plot_correlation(data):
    # Loại bỏ các cột không phải dạng số
    numeric_data = data.select_dtypes(include=['number'])
   
    corr = numeric_data.corr()
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr, annot=True, cmap='coolwarm')
    plt.title('Biểu đồ Tương quan')
    plt.show()

    corr_pairs = corr.unstack().sort_values(ascending=False)
    corr_pairs = corr_pairs[corr_pairs != 1]
    
    print("Những cặp biến có độ tương quan cao nhất:")
    print(corr_pairs.head(10))

plot_correlation(df)
```
