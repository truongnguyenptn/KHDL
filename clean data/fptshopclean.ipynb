```python
import pandas as pd
import re

# Đọc dữ liệu từ file CSV vào DataFrame
df = pd.read_csv(r'../crawldata/fptshop.csv')

# Tạo các cột mới để lưu thông tin tách ra từ cột "Title"
df['Brand'] = df['Title'].str.extract(r'(HP|Asus|Lenovo|MSI|Acer|Dell|MacBook|Macbook|Huawei|Gigabyte|Masstel)', flags=re.IGNORECASE)
df['Name'] = df['Title'].str.extract(r'(?:Laptop|MacBook)\s([^/]+)', flags=re.IGNORECASE)




# Lưu DataFrame đã làm sạch vào một tập tin CSV mới
df.to_csv(r'../clean data/fptshopclean.csv', index=False)

# Hiển thị vài dòng đầu tiên của DataFrame
df.head()
```

```python
# Tên các cột cần xóa
columns_to_drop = ['AMD, Ryzen 5, 7520U', '8 GB (1 thanh 8 GB), DDR4', 'SSD 256 GB', 
                   'AMD Radeon Graphics', 'Windows 11 Home Single Language', '1.36 kg', 
                   '32.4 x 21.5 x 1.79 cm', 'Trung Quốc', '2023', 
                   'https://fptshop.com.vn/may-tinh-xach-tay/hp-245-g10-r5-7520u-9h8x8pt']

# Xóa các cột
df = df.drop(columns=columns_to_drop)

df.to_csv(new_csv_path, index=False, encoding='utf-8')
```

```python
df = df.dropna(subset=['Title']) 
df.to_csv(new_csv_path, index=False)
df.isnull().sum() 
```

```python
def convert_price_to_integer(price):
    if isinstance(price, str):
        price = price.replace('₫', '').replace('.', '')
        return int(price)
    else:
        return price

# Áp dụng hàm chuyển đổi cho cột 'Price sale' và 'Price origin'
df['Price sale'] = df['Price sale'].apply(convert_price_to_integer)
df['Price origin'] = df['Price origin'].apply(convert_price_to_integer)

df['Price origin'].fillna(df['Price sale'], inplace=True)

# Lưu DataFrame đã chỉnh sửa vào một tập tin CSV mới
csv_path = r'.../clean data/fptshopclean.csv'
# Tên mới của các cột
new_columns = {
    'Title': 'Title',
    'Price sale': 'Price sale',
    'Price origin': 'Price origin',
    'Màn hình': 'Screen',
    'CPU': 'CPU',
    'RAM': 'RAM',
    'Ổ cứng': 'Storage',
    'Đồ họa': 'Graphics',
    'Hệ điều hành': 'Operating_System',
    'Trọng lượng': 'Weight',
    'Kích thước': 'Dimensions',
    'Xuất xứ': 'Origin',
    'Năm ra mắt': 'Year',
    'urlre""': 'URL',
    'Brand': 'Brand',
    'Name': 'Name',
    'Storage': 'Storage',
    'Screen': 'Screen',
    'Operating_System': 'Operating_System'
}

# Đổi tên các cột
df = df.rename(columns=new_columns)



df.head()
```

```python
import re

split_data = df['CPU'].str.split(',')
df['Brand'] = split_data.str[0].str.strip()
df['CPU_Model'] = split_data.str[1].str.strip()
df['CPU_Code'] = split_data.str[2].str.strip()

split_data = df['Screen'].str.split(',')
df['Screen_Size'] = split_data.str[0].str.strip()
df['Screen_Size'] = df['Screen_Size'].str.extract(r'(\S{0,4})')
df['Resolution'] = split_data.str[1].str.strip()
df['Panel_Type'] = split_data.str[2].str.strip()
df['Refresh_Rate'] = split_data.str[3].str.strip()
df['Brightness'] = split_data.str[4].str.strip()


df['Refresh_Rate'] = df['Refresh_Rate'].str.extract(r'(\S{0,3})')
df['RAM'] = df['RAM'].str.extract(r'(\S{0,3})')
df['Weight'] = df['Weight'].str.extract(r'(\S{0,3})')
df['Year'] = df['Year'].str.slice(start=-4)

df.head()

df.to_csv(new_csv_path, index=False)
```

```python
split_data = df['Name'].str.split(' ')
df['Brand'] = split_data.str[0].str.strip()
df['Product'] = split_data.str[2:].str.join(' ').str.strip()
df.head()
```

```python
import re
pattern = r'(\d+)\s*(?:TB|GB)'
def extract_storage_capacity(text):
    match = re.search(pattern, str(text))
    if match:
        value = int(match.group(1))
        unit = text.upper().strip()
        if 'TB' in unit:
            value = 1024  
        return value
    else:
        return None

# Apply the function to the 'Hard drive' column
df['Storage'] = df['Storage'].apply(extract_storage_capacity)

df.to_csv(new_csv_path, index=False)
```

```python
def extract_brand(name):
    if isinstance(name, str) and name.startswith('Laptop'):
        return name.split()[1]
    else:
        return 'Mac'


df['Brand'] = df['Title'].apply(extract_brand)

def update_os(row):
    if row['Brand'] == 'Mac':
        return 'Mac'
    else:
        return 'Windows 11'
df['Operating_System'] = df.apply(update_os, axis=1)

df.rename(columns={'Screen_Size': 'Screen size'}, inplace=True)

df.rename(columns={'RAM': 'Ram'}, inplace=True)

df.rename(columns={'Operating_System': 'Operating system'}, inplace=True)

def extract_cpu_info(text):
    if isinstance(text, str):
        parts = text.split(", ")
        if len(parts) >= 3:
            return parts[1] 
    return None

# Áp dụng hàm rút trích thông tin CPU cho cột CPU và lưu vào cột mới 'CPU Type'
df['CPU'] = df['CPU'].apply(extract_cpu_info)
df.to_csv(new_csv_path, index=False)
```

```python
cpu_pattern = r'(?i)i\d+'
def extract_cpu_type(text):
    match = re.search(cpu_pattern, str(text))
    if match:
        return match.group(0)
    return text


# Áp dụng hàm rút trích thông tin CPU cho cột CPU type và lưu vào cột mới 'CPU'
df['CPU'] = df['CPU'].apply(extract_cpu_type)

df.to_csv(new_csv_path, index=False)
```
