```python
import pandas as pd

# Định nghĩa từ điển dịch tên cột từ tiếng Việt sang tiếng Anh
translation_dict = {
    'Loại card đồ họa': 'Graphics card type',
    'Loại CPU': 'CPU type',
    'Dung lượng RAM': 'RAM capacity',
    'Loại RAM': 'RAM type',
    'Số khe ram': 'Number of RAM slots',
    'Ổ cứng': 'Hard drive',
    'Chất liệu': 'Material',
    'Bảo mật': 'Security',
    'Hãng sản xuất': 'Manufacturer',
    'Màn hình cảm ứng': 'Touchscreen',
    'Chất liệu tấm nền': 'Panel material',
    'Tần số quét': 'Refresh rate',
    'Kích thước màn hình': 'Screen size',
    'Công nghệ màn hình': 'Screen technology',
    'Độ phân giải màn hình': 'Screen resolution',
    'Loại màn hình': 'Screen type',
    'Webcam': 'Webcam',
    'Hệ điều hành': 'Operating system',
    'Wi-Fi': 'Wi-Fi'
}

df.rename(columns=translation_dict, inplace=True)

# Lưu DataFrame đã chỉnh sửa vào một tệp CSV mới
df.to_csv(r'..\clean data\cellphoneSclean.csv', index=False)

# In ra tên các cột mới
print(df.columns)

```

```python

df = df.drop(columns='Webcam')
df = df.drop(columns='Screen type')
df = df.drop(columns='Wi-Fi')
df = df.drop(columns='Graphics card type')
df = df.drop(columns='Touchscreen')
df = df.drop(columns='Refresh rate')
df = df.drop(columns='Screen technology')
```

```python
def extract_brand(name):
    if isinstance(name, str) and name.startswith('Laptop'):
        return name.split()[1]
    else:
        return 'Mac'


df['Brand'] = df['Title'].apply(extract_brand)

df.to_csv(r'..\clean data\cellphoneSclean.csv', index=False)

df.head()
```

```python
import numpy as np

df['Price sale'] = df['Price sale'].str.replace('đKhi thu cũ lên đời', '').str.replace('.', '').astype(float)
df['Price origin'] = df['Price origin'].str.replace('đ', '').str.replace('.', '').astype(float)

sale_mean = df['Price sale'].mean()
df['Price sale'].fillna(sale_mean, inplace=True)

df['Price origin'].fillna(df['Price sale'], inplace=True)

df['Price sale'] = df['Price sale'].astype(int)
df['Price origin'] = df['Price origin'].astype(int)

df.to_csv(r'..\clean data\cellphoneSclean.csv', index=False)
```

```python
df = df.dropna(subset='RAM capacity')

df['Screen size'] = df['Screen size'].str.extract(r'(\S{0,4})')
df['Screen size'].fillna('14', inplace=True)

def update_os(row):
    if row['Brand'] == 'Mac':
        return 'Mac'
    else:
        return 'Windows 11'
df['Operating system'] = df.apply(update_os, axis=1)
```

```python
import re
pattern = r'(\d+)\s*(?:TB|GB)'

# Function to extract storage capacity
def extract_storage_capacity(text):
    match = re.search(pattern, str(text))
    if match:
        value = int(match.group(1))
        unit = text.upper().strip()
        if 'TB' in unit:
            value = 1024  
        return value
    else:
        return None

df['Storage'] = df['Hard drive'].apply(extract_storage_capacity)
df['Ram'] = df['RAM capacity'].apply(extract_storage_capacity)
df = df.drop(columns='Hard drive')
df = df.drop(columns='RAM capacity')
df.to_csv(r'..\clean data\cellphoneSclean.csv', index=False)

df.head()
```

```python
cpu_pattern_intel_amd = r'(?i)(Intel Core i\d+|AMD Ryzen \d+|Core i\d+|Intel Core Ultra \d+|Intel® Core™ \S+|\w+-\d+U)'
cpu_pattern_apple = r'(?i)Apple M(\d+)\b'

# Hàm để rút trích thông tin về CPU từ văn bản
def extract_cpu_type(text):
    match = re.search(cpu_pattern_intel_amd, str(text))
    if match:
        cpu_info = match.group(0)
        if 'Intel Core Ultra' in cpu_info:
            return 'Ultra ' + cpu_info.split()[-2]
        elif 'Intel' in cpu_info:
            if 'Core Ultra' in cpu_info:
                return 'i' + cpu_info.split()[-2][1:]
            elif 'Core™' in cpu_info:
                return 'i' + cpu_info.split()[-1][1:]
            elif 'Core' in cpu_info:
                return 'i' + cpu_info.split()[-1][1:]
        elif 'AMD Ryzen' in cpu_info:
            return 'Ryzen ' + cpu_info.split()[-1]
        else:
            return cpu_info
    else:
        match = re.search(cpu_pattern_apple, str(text))
        if match:
            return 'M' + match.group(1)
    return None


# Áp dụng hàm rút trích thông tin CPU cho cột CPU type và lưu vào cột mới 'CPU'
df['CPU'] = df['CPU type'].apply(extract_cpu_type)

# Lưu DataFrame đã chỉnh sửa vào một tệp CSV mới
df.to_csv('../clean data/cellphoneSclean.csv', index=False)
```
